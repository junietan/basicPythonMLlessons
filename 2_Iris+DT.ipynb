```python
# import the relevant packages
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

### The Iris dataset
iris-machinelearning.png

```python
# We can load the iris dataset straight from sklearn
iris = load_iris()
```

```python
# Understanding the iris dataset
print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
print("Shape of data:", iris.data.shape)
print("Description:\n", iris.DESCR)
```

```python
# load data and target from iris dataset
X, y = iris.data, iris.target

# Check what the input looks like
X
```

```python
# Check what the target looks like
y
```

```python
# We see that there are 150 samples, each with 4 feauters
np.shape(X)
```

### Creating the decision tree
First, we need to define the decision tree and its parameters, if any

Then, we need to train/create the tree based on the data

Both are easily achieved through sklearn, with 2 simple commands



```python
# Defining the tree classifier
clf = DecisionTreeClassifier()
```

```python
# Training/creating the decision tree
clf = clf.fit(X, y)
```

```python
# At this point, we have created a fully working decision tree for the Iris dataset
# We can now give this tree an input, and it will predict the class of the flower
prediction=clf.predict([[6.1, 2.7, 3.9, 1.2]])
iris.target_names[prediction]
```

### Visualizing the tree

With sklearn, we also have capabilities to plot the tree. The plot shows how the decision tree makes classifications based on the features.

```python
# The default plot_tree function returns a list with text information about the tree and also plots a small image of it
plot_tree(clf)
```

```python
# To better visualize it, we can use matplotlib, and control the size of the
# figure and add features and labels in the nodes

plt.figure(figsize=(15,12))
plot_tree(clf, filled=True,
feature_names=iris.feature_names,
class_names=iris.target_names)
plt.show()
```

```python
# As can be seen from the image above, though, the features and classes don't have names
# It is not informative, as we don't know what feature X[2] corresponds to
# So, we can add a list of feature and class names to the function
```

Notice that all the leaf nodes (the nodes at the bottom of the tree) have a Gini impurity of 0. This means that each leaf node contains samples of only one class. While this indicates perfect classification on the training data, it's important to be aware that this can be a sign of **overfitting**, especially in a complex tree. Overfitting occurs when the model learns the training data too well, including noise, and may not perform as well on new, unseen data. For training purposes, this perfectly classified tree helps illustrate the decision-making process.

To assess how well our model generalizes to new data and to help prevent overfitting, it's crucial to evaluate it on data it hasn't seen during training. This is where the concept of **splitting the dataset into training and testing sets** becomes important. In the next section, we will explore how to split our data and use these sets to get a more reliable evaluation of our decision tree model.
